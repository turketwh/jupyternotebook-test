```python
# Bring in the plotting library
# Using "as" lets you reference that library via another n
import matplotlib.pyplot as plotter

# Set the names of the entities you want to show in the pie chart
states = ['Freshman','Sophomore','Junior','Senior']
# Set the counts for each associated entity (1 Freshman, 3 Sophomores, ...)
sizes = [1,3,9,12]
# Make the pie chart
plotter.pie(sizes,labels=states,autopct='%1.1f%%')
plotter.axis('equal')
# Give the chart a title
plotter.title('Student Class Distribution')

# Save the pie chart to a png file
# Uncomment if desired
#figure = plotter.gcf()
#figure.savefig('class.png', dpi=300)

#Show the figure
plotter.show()
```
